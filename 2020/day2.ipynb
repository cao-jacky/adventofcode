```python
import numpy as np
```

```python
# read inputs and create list/array
passwords_list = []
day2_inputs = "inputs/day2.txt"
with open(day2_inputs) as file:
    passwords_list = file.readlines()
```

```python
# part 1

# parse through the list and check whether each password is valid

total_valid = 0

for i in range(len(passwords_list)):
    curr_password = passwords_list[i]
    
    cp_split = curr_password.split(":")
    pw_condition = cp_split[0]
    pw_password = cp_split[1]
    
    pwc_split = pw_condition.split(" ")
    pwc_range = pwc_split[0].split("-")
    pwc_letter = pwc_split[1]
    
    pwc_min = int(pwc_range[0])
    pwc_max = int(pwc_range[1])
    
    pwcl_occurence = pw_password.count(pwc_letter)
    
    if pwc_min <= pwcl_occurence <= pwc_max:
        total_valid += 1
        
print(total_valid)
```
