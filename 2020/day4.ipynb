```python
import numpy as np
import json
import re
```

```python
day4_file = [] 
input_file = "inputs/day4.txt"
with open(input_file) as file:
    day4_file = file.readlines()
# print(day4_file)
```

```python
# part 1
passport_data = []
tmp_passport = []
for i in range(len(day4_file)):
    curr_row = day4_file[i]    
    
    if not curr_row == "\n":
        tmp_passport.append(curr_row)
        if i+1 == len(day4_file):
            passport_data.append(tmp_passport)
    else:
        passport_data.append(tmp_passport)
        tmp_passport = []
# print(passport_data)
```

```python
# tidy up the passport data 
ppd_tidied = []
for i in range(len(passport_data)):
    curr_passport = passport_data[i]
    tmp_ppd = []
    for j in range(len(curr_passport)):
        curr_item = curr_passport[j].strip()
        ci_split = curr_item.split(" ")
        for cis_key, cis_value in enumerate(ci_split):
            tmp_ppd.append(cis_value)
    ppd_tidied.append(tmp_ppd)
# print(ppd_tidied)
```

```python
# convert list of passports to dictionary of passports
# then JSON of passports
output = []
for pp_key, pp_value in enumerate(ppd_tidied):
    for key_value in pp_value:
        key, value = key_value.split(':', 1)
        if not output or key in output[-1]:
            output.append({})
        output[-1][key] = value

pp_sorted = sorted(output, key=lambda d: list(d.keys()))
pp_json = json.dumps(pp_sorted, indent = 4)   
pp_json_object = json.loads(pp_json)
```

```python
# analyse and verify the passport data
# check if each passport has eight fields
# if 7 fields, check if only "cid" is missing

# print(pp_json_object)

part1_valid_pp = []

num_valid_pp = 0
for pp_key, pp_value in enumerate(ppd_tidied):
    ppd_len = len(pp_value)
#     print(pp_key, pp_value)
    
    if ppd_len == 8:
        num_valid_pp += 1
        part1_valid_pp.append(pp_value)
    if ppd_len == 7:
        if not any("cid" in pp_field for pp_field in pp_value):
#             print(pp_value)
            num_valid_pp += 1
            part1_valid_pp.append(pp_value)
print(num_valid_pp)
```

```python
# part 2
# data validation needs to be performed

# validation rules:
val_rules = {
    "byr": [1920, 2002],
    "iyr": [2010, 2020],
    "eyr": [2020, 2030],
    "hgt": {
        "cm": [150, 193],
        "in": [59, 76]
    },
    "hcl": 6,
    "ecl": ["amb", "blu", "brn", "gry", "grn", "hzl", "oth"],
    "pid": 9,
    "cid": 0
}

total_valid = 0

for key_vp, value_vp in enumerate(part1_valid_pp):
    curr_pp = value_vp
    
    cpp_fields_valid = 0
    for i in range(len(curr_pp)):
        curr_field = curr_pp[i]
        
        cf_split = curr_field.split(":")
        field_name = cf_split[0]
        field_value = cf_split[1]
        
        field_cond = val_rules[field_name]
#         print(field_name, field_cond)
        
        if "pid" in field_name:
            check_cond = len(list(field_value))
            if check_cond == 9:
                cpp_fields_valid += 1
        if "ecl" in field_name:
            if field_value in field_cond:
                cpp_fields_valid += 1
        if "hcl" in field_name:
            if "#" in field_value[0]:
                colour_code = field_value[1:]
                if len(colour_code) == 6:
                    cpp_fields_valid += 1
        if "hgt" in field_name:
            match = re.match(r"([0-9]+)([a-z]+)", field_value, re.I)
            if match:
                items = match.groups()
                
            height_val = items[0]
            height_unit = items[1]
            
            height_cond = field_cond[height_unit]
            
            if height_cond[0] <= int(height_val) <= height_cond[1]:
                cpp_fields_valid += 1
                
        if field_name in ["byr", "iyr", "eyr"]:
            if field_cond[0] <= int(field_value) <= field_cond[1]:
                cpp_fields_valid += 1
#         if "cid" in field_name:
#             cpp_fields_valid +=1
    
#     print(cpp_fields_valid)
    if cpp_fields_valid in [7]:
        total_valid += 1
print(total_valid)
            
#             else:
#                 print("field not valid", curr_field)
```

```python
import sys
import re

passports = []
i = 0
passports.append({})
with open("inputs/day4.txt") as input_file:
    for line in input_file:
        line = line.strip()
        if line == "": 
            i += 1
            passports.append({})
            continue

        passports[i].update(dict(x.split(":") for x in line.split(" ")))

validPassport = 0
requiredFields = ('byr', 'iyr', 'eyr', 'hgt', 'hcl', 'ecl', 'pid')

for passport in passports:
    if not all ( k in passport for k in requiredFields): continue
    if not (int(passport['byr']) >= 1920 and int(passport['byr']) <= 2002): continue
    if not (int(passport['iyr']) >= 2010 and int(passport['iyr']) <= 2020): continue
    if not (int(passport['eyr']) >= 2020 and int(passport['eyr']) <= 2030): continue
    if not passport['ecl'] in ("amb", "blu", "brn", "gry", "grn", "hzl", "oth"): continue
    if not re.fullmatch("#[0-9a-f]{6}", passport['hcl']): continue
    if not re.fullmatch("\d{9}", passport['pid']): continue
    if not re.fullmatch("(\d{2,3})(cm|in)", passport['hgt']): continue

    height = re.match("^(\d{2,3})(cm|in)$", passport['hgt']).groups()
    if height[1] == "in":
        if not (int(height[0]) >= 59 and int(height[0]) <= 76): continue
    if height[1] == "cm":
        if not (int(height[0]) >= 150 and int(height[0]) <= 193): continue
    
#     print(passport)
    validPassport += 1


print("Valid passports: ", validPassport)
```
